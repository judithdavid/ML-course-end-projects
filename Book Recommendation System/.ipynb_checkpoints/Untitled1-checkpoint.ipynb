```python
rating = "BX-Book-Ratings.csv"
book = "BX-Books.csv"
users = "BX-Users.csv"
```

```python
r = pd.read_csv(rating,encoding="latin-1")
b = pd.read_csv(book,encoding="latin-1")
u = pd.read_csv(users,encoding="latin-1")
```

```python
# 3. Read the data where ratings are given by users
```

```python
# merging the two data set books and rating.
```

```python
df=pd.merge(r,b,on='isbn')
```

```python
df
```

```python
# 4. Convert the user_id variable to numeric numbers in the correct order.
# 5. Convert both user_id and ISBN to the ordered list, i.e., from 0...n-1
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE = LabelEncoder()
# 'user_id_order', 'isbn_order'
```

```python
df['isbn_order']=LE.fit_transform(df['isbn'])
```

```python
df['user_id_order']=LE.fit_transform(df['user_id'])
```

```python
df
```

```python
# Re-index the columns to build a matrix.
```

```python
ordered_col = ['user_id_order', 'isbn_order', 'rating', 'book_title', 'book_author', 'year_of_publication','publisher',
               'user_id', 'isbn']
df = df.reindex(columns = ordered_col)
```
